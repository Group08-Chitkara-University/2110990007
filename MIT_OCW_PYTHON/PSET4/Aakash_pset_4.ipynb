```python
#Caesar cipher 

my_word = "AMAZON"
shift = int(input())


def cipher(my_word , shift):
    l = []
    for i in range(len(my_word)):
        l.append(my_word[i])
    
    for j in range(len(l)):
        l[j] = chr(ord(l[i])+shift)
    
    for k in l:
        print(k,end="")
cipher(my_word , shift)
```
